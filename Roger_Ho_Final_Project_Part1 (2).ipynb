Because I realized how code is very similar and to follow DRY principle, I created a general method for plotting related to counting a certain data columns values.

AND THUS, I WENT BACK TO USING THIS METHOD TO PLOT THE PRIVOUS VISUALIZATIONS WITHOUT REMOVING MY PREVIOUS CODE

```python
def get_x_y_for_plotting(pandas_series):
    count_map = {}
    for colum_data in pandas_series:
        count_map[colum_data] = count_map.get(colum_data, 0) + 1

    colum_data_list = []
    count_list = []
    for colum_data, count in count_map.items():
        colum_data_list.append(colum_data)
        count_list.append(count)
    return colum_data_list, count_list
```

```python
x, y = get_x_y_for_plotting(df[df.arrestee_sex.isin(['MALE', 'FEMALE'])].arrestee_sex)
plt.bar(x, y)
plt.title("Arrested Counts Associated to Gender in Urbana From 1989 ~ 2018")
plt.xlabel('Gender')
plt.ylabel("Arrested Counts")
```
